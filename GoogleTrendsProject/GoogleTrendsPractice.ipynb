### Notebook is pulling out the trend data for the list of mental diseases

Utilizing pytrend library


```python
import pandas as pd
import time
```

```python
from pytrends.request import TrendReq
```

```python
# connecting to the Google
pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), requests_args={'verify':False})
```

```python
# building the Time frames
time_period_min_list = ["2004-01-01","2005-01-01","2006-01-01","2007-01-01","2008-01-01","2009-01-01","2010-01-01","2011-01-01","2012-01-01","2013-01-01","2014-01-01","2015-01-01","2016-01-01","2017-01-01","2018-01-01","2019-01-01","2020-01-01","2021-01-01","2022-01-01"]
time_period_max_list = ["2004-12-31","2005-12-31","2006-12-31","2007-12-31","2008-12-31","2009-12-31","2010-12-31","2011-12-31","2012-12-31","2013-12-31","2014-12-31","2015-12-31","2016-12-31","2017-12-31","2018-12-31","2019-12-31","2020-12-31","2021-12-31","2022-12-31"]
```

```python
frames = [i[0]+" " +i[1] for i in list(zip(time_period_min_list, time_period_max_list))]
```

```python
# list of the keywords for the search
kw_list = ['depression', 'anxiety' ,'add', 'bipolar disorder', 'autism']
```

```python
pytrends.build_payload(kw_list, cat=0, timeframe='all', geo='US', gprop='')
```

```python
# trends for all available years in usa
pytrends.build_payload(kw_list, timeframe='all', geo = 'US')
data = pytrends.interest_over_time()
```

```python
data.head()
```

```python
data.to_csv('interestOverAllTimeUSA.csv')
```

```python
# Pulling data by the regions within the USA

# within the usa lets look into interest by the regions by the years
# that where we will use the frame
lst =[]
for i in frames:
  pytrends.build_payload(kw_list, timeframe=i, geo = 'US')
  regions = pytrends.interest_by_region(resolution = 'region' )
  regions['date'] = i
  lst.append(regions)
  time.sleep(4)
```

```python
# saving the dataframe
AllRegionsYear = pd.concat(lst)
AllRegionsYear.to_csv('TrendsByUSRegionsAllYears.csv')
```

```python
AllRegionsYear.head()
```

```python
# trend by the countries

# need to reconnect to pytrends so I could do countries
pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), requests_args={'verify':False})
countryList =[]

for i in frames:
  pytrends.build_payload(kw_list, timeframe= i, geo='')
  countries = pytrends.interest_by_region(resolution='COUNTRY')
  countries['date'] = i
  countryList.append(countries)
  time.sleep(1)
```

```python
ContriesInterestAllYears = pd.concat(countryList)
ContriesInterestAllYears.to_csv('CountriesInterestALLYears.csv')
```
